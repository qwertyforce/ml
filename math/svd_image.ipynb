```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = np.array(Image.open("./test.jpg").convert('L'))
U,S,VT = np.linalg.svd(img,full_matrices=False)
S=np.diag(S)
```

```python
r = 100 #rank
approx = U[:,:r] @ S[:r,:r] @ VT[:r,:]
```

```python
plt.set_cmap("gray")
plt.imshow(256-approx)
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title("Singular values")
plt.show()

plt.figure(2)
plt.semilogy(np.cumsum(np.diag(S)/np.sum(np.diag(S))))
plt.title("Singular values cum_sum")
plt.show()
```
