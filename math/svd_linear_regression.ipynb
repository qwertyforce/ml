```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3 # True slope
a=np.arange(-2,2,0.25).reshape(-1,1)
b= x*a + np.random.rand(*a.shape)

#Ax=b
#x = A_inv * b 
U, S, VT = np.linalg.svd(a, full_matrices=False)
S = np.diag(S)
A_pseud_inv =  VT.T@np.linalg.inv(S)@U.T
x_approx =A_pseud_inv@b
```

```python
plt.plot(a,x*a,label="True line")
plt.plot(a,b,"x",label="Noisy data")
plt.plot(a,x_approx*a,"--",label="Least-square fit")
plt.legend()
plt.show()
```
