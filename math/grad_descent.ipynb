```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(-10,10,1000) #we generated 1000 point between -10 and 10

y = x**2 - 15*x + 5 
fig, ax = plt.subplots()
#fig.show()

def grad(x):
    return 2*x-15

lr = 0.2
w=-100
for i in range(30):
    gradient = grad(w)
    w -= lr*gradient 
    print(w)
ax.plot(x,y)
ax.plot(w,(w**2 - 15*w + 5 ), "s")
```
